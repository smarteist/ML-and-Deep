```python
from statsmodels.formula import api as smf
from statsmodels import api as sm
from pandas import pandas as pd
```

```python
data = pd.read_csv('3campain.csv')
```

```python
data.head()
```

```python
result = smf.ols("Sell~Campain", data=data).fit()
```

```python
sm.stats.anova_lm(result)
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Data for the correlation table
data = {
    "SWP": [5 , 2 , 6 , 15, 25, 2, 4, 62],
    "HWP": [4 , 4 , 2 , 7 , 3 , 0, 1, 2 ],
    "SEM": [1 , 2 , 3 , 10, 19, 8, 7, 10],
    "CNF": [1 , 5 , 7 , 1 , 13, 1, 6, 25],
    "DEP": [1 , 6 , 6 , 5 , 7 , 0, 6, 27],
    "MEM": [2 , 4 , 3 , 3 , 3 , 0, 1, 8 ],
    "CON": [1 , 4 , 3 , 3 , 3 , 2, 1, 4 ],
    "EC":  [1 , 1 , 1 , 7 , 8 , 0, 2, 14],
    "TM":  [5 , 2 , 11, 11, 13 ,0, 4, 5 ],
}

# Labels for the rows
labels = [
    "Device: Hardware",
    "Device: Firmware",
    "Communication: Connectivity",
    "Communication: Messaging",
    "Cloud: DM",
    "Cloud: Automation",
    "Compatibility",
    "DEV",
]



# Descriptions for each root cause (data labels)
descriptions = {
    "SWP": "Software programming faults",
    "HWP": "Hardware programming faults",
    "SEM": "Semantic programming faults",
    "CNF": "Configuration faults",
    "DEP": "Dependency faults",
    "MEM": "Memory faults",
    "CON": "Concurrency faults",
    "EC": "Exceptional case faults",
    "TM": "Timing faults",
}

# Create a DataFrame
df = pd.DataFrame(data, index=labels)

# Create a list of full column names with descriptions
column_descriptions = [f"{col} - {descriptions[col]}" for col in df.columns]

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df, annot=True, cmap="Blues", fmt="d", cbar=False, annot_kws={"fontsize": 10})

# Add descriptive labels to the axes
plt.xticks(ticks=np.arange(len(column_descriptions)) + 0.5, labels=column_descriptions, rotation=45, ha="right", fontsize=10)
plt.yticks(rotation=0, fontsize=10)


# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
